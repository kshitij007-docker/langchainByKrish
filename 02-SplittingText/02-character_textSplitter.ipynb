```python

from langchain.text_splitter import CharacterTextSplitter
speech="" ##initializing an empty string variable named speech.
with open('speechKs.txt', 'r') as file: ##open function is used to read a file from the directory.It is python's built-in function.here 'r' means read mode.and file is the file object.
    speech=file.read() ##The read() method reads the entire content of the file as a single string.
speech ##Now the variable speech contains the entire content of the file as a single string.

text_splitter=CharacterTextSplitter(chunk_size=100,chunk_overlap=20) ##CharacterTextSplitter is a class from the langchain library that is used to split text into smaller chunks based on character count.
##It is different from recursive character text splitter. as it does not use recursion to split the text.Therefore it is faster than recursive character text splitter.
text=text_splitter.create_documents([speech]) 
print(text)
print(text[0])
print (text[1])
```

```python
from langchain.text_splitter import CharacterTextSplitter

with open('SpeechKs.txt') as f:
    speech = f.read()

    text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)
    text=text_splitter.create_documents([speech])
    print(text[0])
    print(text[1])
```
