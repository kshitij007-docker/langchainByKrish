```python
from langchain_text_splitters import HTMLHeaderTextSplitter 

url= "https://plato.stanford.edu/entries/plato/"


headers_to_split_on=[("h1","Header 1"),("h2","Header 2")] ##here headers_to_split_on is used so that we can split html page as per headers that is h tag
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on) ##Here html_splitter will carry the value return by htmlHeaderTextSplitter
html_headers_splits=html_splitter.split_text_from_url(url) ##Here we have provided the source of html page and store the data in html_headers_splits
html_headers_splits

```
