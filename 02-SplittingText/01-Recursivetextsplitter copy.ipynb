```python
## Reading a PDF File

from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('eu.pdf')
docs=loader.load()
docs
```

```python
###How to recursively split text by characters

from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs) ##here split documents is a method of the RecursiveCharacterTextSplitter class that takes a list of documents as input and splits them into smaller chunks based on the specified chunk size and overlap.
##we have not used create_documents method here because docs is already a list of Document objects.which means we can directly pass it to the split_documents method.
final_documents
```

```python
####Reading a text file and splitting it into chunks
from langchain_community.document_loaders import TextLoader
loader=TextLoader('speechKs.txt')
docs=loader.load()
docs
```

```python
speech="" ##initializing an empty string variable named speech.
with open('speechKs.txt', 'r') as file: ##open function is used to read a file from the directory.It is python's built-in function.here 'r' means read mode.and file is the file object.
    speech=file.read() ##The read() method reads the entire content of the file as a single string.
speech ##Now the variable speech contains the entire content of the file as a single string.

text_splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20) ##Here we are creating an instance of the RecursiveCharacterTextSplitter class with a chunk size of 100 characters and an overlap of 20 characters between chunks.
text=text_splitter.create_documents([speech]) ##here we are using the create_documents method of the text_splitter object to split the speech text into smaller chunks based on the specified chunk size and overlap.
##create_documents method takes a list of strings as input and returns a list of Document objects, where each Document object contains a chunk of text.
##We are passing [speech] as a list containing a single string (the entire speech text) to the create_documents method.
##Now text variable contains a list of Document objects, each representing a chunk of the original speech text.
print(text)
print(text[0])
print (text[1])
```
