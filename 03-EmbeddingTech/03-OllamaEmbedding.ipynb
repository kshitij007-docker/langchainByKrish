```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(OllamaEmbeddings(model="gemma3:1b") ) ##By default model it uses is LLama2. Now we have defined the model to be gemma3:1b. Here we have intialized the embedding model.
```

```python
embeddings ##Here we can see the details of the model we have used.
```

```python
r1=embeddings.embed_documents(["A",
                               "B"]) ##Here we are embedding two documents.
r1
```

```python
len(r1[0]) ##Here we can see the length of the embedding vector of first document.
len(r1[1])
```

```python
embeddings.embed_query("What is the second letter of Greek alphabet?")
```

```python
##Other embedding models : https://ollama.com/blog/embedding-models

embeddings= OllamaEmbeddings(model="mxbai-embed-large") ##Here we have changed the embedding model to mxbai-embed-large.
text="This is test document."
query_result=embeddings.embed_query(text) ##Here we are embedding a query.
query_result ##Here we can see the embedding vector of the query mentioned above.It will return a vector of length 1536.
```
