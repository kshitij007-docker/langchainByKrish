```python
import os ## it is used to set environment variables. which means we can set the HF_TOKEN environment variable to the value in .env file
from dotenv import load_dotenv ##to load the .env file
from langchain_huggingface import HuggingFaceEmbeddings ##this is the module to import HuggingFace Embedding model from langchain. it will be used to create the embedding object which will be used to embed the text.
load_dotenv() #load the .env file
os.environ['HF_TOKEN']=os.getenv("HF_TOKEN") #set the HF_TOKEN environment variable to the value in .env file
embeddings = HuggingFaceEmbeddings(
	model_name="sentence-transformers/all-MiniLM-L6-v2" ## you can use any model from HuggingFace.This model is small and fast to embed text and sentences.
)
```

```python
text="this is a sample text"
query_result=embeddings.embed_query(text)
print(query_result)
len(query_result)
```

```python
doc_result=embeddings.embed_documents([text,"this is just another text"])
doc_result[0]
```
