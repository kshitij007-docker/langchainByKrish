```python
##Step -1 Text Loader

from langchain_community.document_loaders import TextLoader
loader=TextLoader('speechKs.txt')
loader
```

```python
### Step 2:recursively split text by characters

from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(loader) ##here split documents is a method of the RecursiveCharacterTextSplitter class that takes a list of documents as input and splits them into smaller chunks based on the specified chunk size and overlap.
##we have not used create_documents method here because docs is already a list of Document objects.which means we can directly pass it to the split_documents method.
final_documents
```

```python
### Step 3: Embedding with OpenAI
import os
from dotenv import load_dotenv
from langchain_openai import OpenAIEmbeddings
load_dotenv() # load all my environment variables

os.environ["OPEN_API_KEY"]=os.getenv("OPENAI_API_KEY") # Set environment variable for OpenAI API key.It is to load env variable from .env file and set it to os.environ for use in the script.

embedding_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)# Initialize OpenAIEmbeddings with 1024 dimension model.
embedding_1024
```

```python
### Step 4: Store Embedded Vectors in ChromaDB
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_documents,embedding_1024) # final_documents is the list of documents which is split into chunks , check above variable and embedding_1024 is the embedding model which is used to convert the text into vector.
db
```

```python
##Lets query db and retreive some data using similarity search

query ="also known as Sam Bahadur" # it is just random string from speechKs.txt file
retrieved_results=db.similarity_search(query)
print(retrieved_results) # it returns the list of documents which are similar to the query.
```
