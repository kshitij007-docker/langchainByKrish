```python
## Load the properties file

import os
from dotenv import load_dotenv
from langchain_community.llms import Ollama
import streamlit as st
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

## Langsmith Tracking

os.environ['LANGCHAIN_API_KEY']=os.getenv("LANGCHAIN_API_KEY")
os.environ['LANGCHAIN_TRACING_V2']="true"
os.environ['LANGCHAIN_PROJECT']=os.getenv("LANGCHAIN_PROJECT")
```

```python
## Prompt Template. It is used to provide the context in the prompt.

prompt= ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant.Please respond to the question asked"),
    ("user","Question: {question}") ##Here question is the input from the user.
])
```

```python
## Streamlit framework is used to create the web app

st.title("langchain demo with Ollama LLM model gemma3:1b") ## Title of the web app
input_text =st.text_input("Enter your first question here:") ## Input text box to get the question from the user

```

```python
## call ollama gemma3:1b 

llm =Ollama(model="gemma3:1b") 
output_parser=StrOutputParser() ## Output parser to parse the output from the llm which means to convert the output into string format.
chain=prompt|llm|output_parser ## Create the chain by combining the prompt, llm and output parser.

if input_text: ## Check if the input text is not empty
   st.write(chain.invoke({"question":input_text})) ## Invoke the chain with the input question from the user and display the output in the web app.
```
